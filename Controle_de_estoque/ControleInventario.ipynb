# Controle de Estoque  Cultura Pura 

```python
class Insumo:
    def __init__(self, codigo, nome, quantidade=0):
        self.codigo = codigo
        self.nome = nome
        self.quantidade = quantidade

class Fase:
    def __init__(self, nome, insumos):
        self.nome = nome
        self.insumos = insumos

class Fermentacao:
    def __init__(self, tipo):
        self.tipo = tipo
        self.fases = []

    def adicionar_fase(self, nome_fase, insumos):
        fase = Fase(nome_fase, insumos)
        self.fases.append(fase)

    def listar_fases(self):
        for fase in self.fases:
            print(f'Fase: {fase.nome}')
            for insumo, quantidade in fase.insumos.items():
                print(f'    Insumo: {insumo}, Quantidade: {quantidade}')

class Estoque:
    def __init__(self):
        self.insumos = []
        self.fermentacoes = []

    def cadastrar_insumo(self, codigo, nome, quantidade=0):
        insumo = Insumo(codigo, nome, quantidade)
        self.insumos.append(insumo)

    def cadastrar_insumos(self, lista_insumos):
        for insumo in lista_insumos:
            self.cadastrar_insumo(insumo['codigo'], insumo['nome'], insumo['quantidade'])

    def adicionar_quantidade_insumo(self, nome, quantidade):
        for insumo in self.insumos:
            if insumo.nome == nome:
                insumo.quantidade += quantidade

    def cadastrar_fermentacao(self, tipo, fases):
        fermentacao = Fermentacao(tipo)
        for nome_fase, insumos in fases.items():
            fermentacao.adicionar_fase(nome_fase, insumos)
        self.fermentacoes.append(fermentacao)

    def listar_insumos(self):
        for insumo in self.insumos:
            print(f'Código: {insumo.codigo}, Nome: {insumo.nome}, Quantidade: {insumo.quantidade}')
```
